# K-Space

## Qué es el K-Space?
El K-space es una **representación de las frecuencias espaciales** de una señal.

En el campo del MRI sucede que una vez adquiridos los datos del FID (Free Induction Delay) y almacenados en una matriz, estos datos se encuentran guardados en función de la frecuencia y fase de oscilación del spin en formato TF por lo que debe aplicarse la TF inversa para obtener la imagen del estudio.

A su vez, se pueden aplicar transformaciones matemáticas en el K-Space para mejorar la calidad de la imagen o acelerar el proceso de adquisición de imágenes.

## Relación entre el K-Space y la TF2D
**Podemos aplicar la Transformada de Fourier para pasar de nuestra imagen al K-Space como tambien la Anti Transformada de Fourier para pasar del K-Space a nuestra imagen**. Esto nos permite afirmar que el K-Space hereda muchas propiedades de las vistas de la TF2D para imagenes: las bajas frecuencias espaciales se mappean en el centro del K-Space dando idea del contraste y la SNR de la imagen y las altas frecuencias espaciales se encuentran en los bordes dando información sobre la resolucion de la imagen.

Si consideramos entonces la frecuencia espacial de las mediciones de la frecuencia como $k_{FE}$ y la frecuencia espacial de las mediciones de la fase como $k_{PE}$, podemos afirmar que el par (x,y) y el par de frecuencias espaciales ($k_{FE}$, $k_{PE}$) admiten la transformacion de Fourier.
